**Filtering Data**

```python
##Grass AND Poison
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
##Grass T1 OR Poison T2
df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
```

```python
##More filters
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP']>70)]
```

```python
##Import regex
import re

##Remove any Megas
df.loc[df['Name'].str.contains('Mega ')]  

##Only pokemon whose type 1 is grass or fire
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I, regex=True)]

##Only pokemon whose name starts with pi
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I, regex=True)]
```

**Conditional Changes**

```python
##Conditionally change a value in a df
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flameo'
df.loc[df['Type 1'] == 'Flameo', 'Type 1'] = 'Fire'
df

##Save df
df.to_csv('modified_2.csv', index=False)
```

```python
##Change  a value in several columns
df.loc[df['Total']>500, ['Generation', 'Legendary']] = ['WOW!', 'Look at that!']
df
```

```python
##Load the original data back in 
df = pd.read_csv('modified_2.csv')
df
```

**Aggregate Statistics (Groupby)**

```python
##Average HP of pokemon by type
df.groupby(['Type 1']).mean().sort_values('Attack',ascending=False)
##Note: you can also do sum or count

##Create a count column then count using groupby(). Display only count column
df['Count'] = 1
df.groupby(['Type 1']).count()['Count'].sort_values(ascending=False)
```

**Working with large amounts of data**

```python
##Create new df with same column names
new_df = pd.DataFrame(columns=df.columns)

##Read a chunk and count the number of types per chunk, append results to new_df
for df in pd.read_csv('modified_2.csv',chunksize=5):
    results = df.groupby(['Type 1']).count()
    ##Append results into the new df
    new_df = pd.concat([new_df, results])
    
new_df
```
